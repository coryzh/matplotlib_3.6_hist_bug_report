```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

print(f"Runs on numpy {np.__version__} and matplotlib {mpl.__version__}")
```

### The scaling issue of y axis

Normally, changes on the scale parameter should only affect the value of x-axis but not the scaling of the y-axis.

```python
%matplotlib inline

scale = 1
data = np.load("test.npy") * scale

hist = plt.hist(data, bins=100, density=True, histtype="step")
```

### Also, for some simulated data

```python
scale = 1.0
test_random = np.random.randn(len(data)) * scale

fig, ax = plt.subplots(1, 2, figsize=(20, 10))
hist_bar = ax[0].hist(test_random, bins=100, density=True, histtype="bar")
hist_step = ax[1].hist(test_random, bins=100, density=True, histtype="step")
```
